```python
#Import Packages
import pandas as mypd
import numpy as mynp
from sklearn.impute import KNNImputer
```

```python
#import Data File of kidney_disease
mydata=mypd.read_csv("C:/Users/demo/Desktop/Data Preprocessing/kidney_disease.csv")
mydata.head(3)
```

```python
mydata.info()
```

```python
#Drop the column ID
mydata=mydata.iloc[:,1:]
#Seprate Object type data type
mydata_object_feat=mydata.select_dtypes(include='object')
mydata_object_feat_name=mydata_object_feat.columns.values
mydata_object_feat_name
```

```python
#Import Encoder
from sklearn.preprocessing import OrdinalEncoder
ode=OrdinalEncoder(dtype=int)
```

```python
#Define Encoding Function
def encode(data):
    #retain non-null values
    nonulls=mynp.array(data.dropna())
    #Convert it into array with one colunm
    nonulls_reshape=nonulls.reshape(-1,1)
    #encode the data
    impute_ordinal=ode.fit_transform(nonulls_reshape)
    #Assignback encoded values to non null values
    data.loc[data.notnull()]=mynp.squeeze(impute_ordinal)
    return(data)
```

```python
#Call the function
for feat in mydata_object_feat_name:
    encode(mydata[feat])
```

```python
mydata.head()
```

```python
#KNN imputer
knn=KNNImputer(n_neighbors=4)
cleandata=knn.fit_transform(mydata)
#Convert to Data FrameObject
cleandata=mypd.DataFrame(cleandata,columns=mydata.columns)
cleandata.head(10)
```

```python
#export the file
cleandata.to_excel("C:/Users/demo/Desktop/Data Preprocessing/kidney_disease_cleaned.xlsx")
```
